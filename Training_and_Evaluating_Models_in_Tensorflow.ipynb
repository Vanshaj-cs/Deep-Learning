```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv("insurance.csv")
```

```python
dataset.head()
```

```python
dataset = pd.get_dummies(dataset)
dataset = dataset.astype(int)
dataset.head()
```

```python
X = dataset.drop("charges", axis = 1)
Y = dataset["charges"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

Traning and improving model

```python
tf.random.set_seed(42)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model.compile(loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.SGD(),
              metrics=['mae'])

model.fit(X_train, Y_train, epochs = 100)
```

```python
model.evaluate(X_test, Y_test)
```

Improving model

```python
tf.random.set_seed(42)

model_2 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model_2.compile(loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.Adam(),
              metrics=['mae'])

model_2.fit(X_train, Y_train, epochs = 100)
```

```python
model_2.evaluate(X_test,Y_test)
```

```python
tf.random.set_seed(42)

model_3 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model_3.compile(loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.Adam(),
              metrics=['mae'])

history = model_3.fit(X_train, Y_train, epochs = 200)
```

```python
model_3.evaluate(X_test, Y_test)
```

```python
pd.DataFrame(history.history).plot()
plt.ylabel("loss")
plt.xlabel("epochs")
```

# **Preprocessing**

```python
 dataset_1 = pd.read_csv("insurance.csv")
```

```python
dataset_1.head()
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
```

```python
ct = make_column_transformer(
    (MinMaxScaler(), ["age", "bmi", "children"]),
    (OneHotEncoder(handle_unknown = "ignore"), ["sex", "smoker", "region"])
)
```

```python
X = dataset_1.drop("charges", axis = 1)
Y = dataset_1["charges"]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
ct.fit(X_train)
X_train_normal = ct.transform(X_train)
X_test_normal = ct.transform(X_test)
```

```python
X_train_normal[0]
```

```python
X_train.shape, X_train_normal.shape
```

```python
tf.random.set_seed(42)

model_4 = tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model_4.compile(loss = tf.keras.losses.mae,
              optimizer = tf.keras.optimizers.Adam(),
              metrics=['mae'])

model_4.fit(X_train_normal, Y_train, epochs = 100)
```

```python
#evaluation
model_4.evaluate(X_test_normal, Y_test)
```
